# Text Preprocessing

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W", " ", text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
X = data['text']
y = data['class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,train_size=0.2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize the TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the training data
xv_train = vectorizer.fit_transform(x_train)

# Transform the test data using the same vocabulary
xv_test = vectorizer.transform(x_test)

```

```python
# Print the feature names (words) in the vocabulary
print("Vocabulary:", vectorizer.get_feature_names_out())
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(xv_train,y_train)
```

```python
pred_lr=lr.predict(xv_test)
```

```python
accuracy_score(y_test,pred_lr)
```

```python
print(classification_report(y_test,pred_lr))
```

# Model Testing

```python
def output_lable(n):
  if n == 0:
    return "Fake News"
  elif n == 1:
    return "Not A Fake News"



def manual_testing(news):
  testing_news = {"text":[news]}
  new_def_test = pd.Dataframe(testing_news)
  new_def_test["text"] = new_def_test["text"].apply(wordopt)
  new_x_test = new_def_test["text"]
  new_xv_test = vectorizer.transform(new_x_test)
  pred_lr = lr.predict(new_xv_test)
  pred_dt = Dt.predict(new_xv_test)
  pred_gbc = Gbc.predict(new_xv_test)
  pred_rf = rf.predict(new_xv_test)

  return print("\n\nlr Prediction: {} \nDt Prediction: {} \nGbc Prediction: {} \nRf Prediction {}".format(output_lable(pred_lr[0]),
                                                                                                          output_lable(pred_gbc[0]),
                                                                                                          output_lable(pred_rf[0])))

```
